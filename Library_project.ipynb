```python
class Library: 
    def __init__(self, name): ### initializing the library
        self.booksList = [] 
        self.name = name
        self.lended_books = []

    def display_available_books(self): # Display the books list 
        print(f"Available books in {self.name} library:")
        for i in self.booksList:
            if i not in [lended_book[1] for lended_book in self.lended_books]: ### Check if the book is not currently lended to a user
                print(f"- {i}") 

    def lend_book(self, user, book):
        if book in [lended_book[1] for lended_book in self.lended_books]:
            print(f"{book} is already lended to another user.") ### check if the book was already lended  
        else:
            self.lended_books.append((user, book))
            print(f"{user} has lended {book}.") ### if the book is not lended add it to the lended list

    def add_book(self, book):
        if book in self.booksList:
            print(f"{book} already exists in the library.") ### check if the book was added already
        else:
            self.booksList.append(book)
            print(f"{book} added to the library.") ### if the book wasnot in the list add the book

    def return_book(self, user, book):
        if (user, book) not in self.lended_books:
            print(f"{book} is not lended by {user}.") ### check if the book was lended by the user 
        else:
            self.lended_books.remove((user, book))
            print(f"{book} returned by {user}.") ### if the book was lended by the user remove it from the lended books

Harry = Library("Harry's Library")
Harry.booksList = ['Python Basics', 'Java Basics', 'SQL Basics', 'C++ Basics', 'R Basics'] ### adding books to the library

while True:
    print("\nHarry's Library:")
    print("1. Display Books")
    print("2. Lend a Book")
    print("3. Add a Book")
    print("4. Return a Book")
    print("5. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        Harry.display_available_books() ### display the books list
    elif choice == 2:
        user = input("Enter user name: ")
        book = input("Enter book name: ")
        Harry.lend_book(user, book) ### the user enter his name and choose the book he want to lend
    elif choice == 3:
        book = input("Enter book name: ")
        Harry.add_book(book) ### add a new book to the list
    elif choice == 4:
        user = input("Enter user name: ")
        book = input("Enter book name: ")
        Harry.return_book(user, book) ### the user enter his name and return the book he lent
    elif choice == 5:
        print("Thank you for using Harry's Library")
        break ### exit the system 
    else:
        print("Invalid choice, please try again.") ### message to the user to choose from the given list
```
